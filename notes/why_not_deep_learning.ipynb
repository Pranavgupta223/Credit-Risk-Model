# 🎯 The Correct Professional Answer

If an interviewer asks:

Why didn’t you use deep learning?

A strong answer would be:

Deep learning typically requires large datasets and complex feature interactions to outperform simpler models. This dataset had only 1000 structured tabular samples with mostly categorical financial variables. In such settings, regularized logistic regression or gradient boosting usually perform better and are more stable. Since logistic regression already achieved strong performance and offered interpretability aligned with regulatory requirements, deep learning would likely add complexity without meaningful gain.

That answer shows:

You understand when DL is appropriate.

You understand dataset size matters.

You understand tabular vs unstructured data.

You understand regulatory context.

You value simplicity when sufficient.

That is mature thinking.
